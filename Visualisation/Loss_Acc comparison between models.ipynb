# Vizualisations

```python
import matplotlib.pyplot as plt

def plot_models(attribute):
  # Plotting test loss
  epochs_big = list(range(0, len(xLSTM_big['test_loss'])))
  epochs_small = list(range(0, len(xLSTM_small['test_loss'])))

  plt.figure(figsize=(10, 6))
  plt.plot(epochs_big, xLSTM_big[attribute], label=f'xLSTM Big - {attribute}', color="g")
  plt.plot(epochs_small, xLSTM_small[attribute], label=f'xLSTM Small - {attribute}', linestyle='--', color="g")
  plt.plot(epochs_small, LSTM_big[attribute], label=f'LSTM Big - {attribute}', color = "r")
  plt.plot(epochs_small, LSTM_small[attribute], label=f'LSTM Small - {attribute}', linestyle='--', color ="r")

  plt.title(f'{attribute} Comparison')
  plt.xlabel('Epoch')
  plt.ylabel(attribute)
  plt.legend()
  plt.grid(True)
  plt.show()
```

```python
for attribute in ["train_loss", "test_loss", "train_accuracy", "test_accuracy"]:
  plot_models(attribute)
```

```python
epochs_big = list(range(0, len(xLSTM_big['test_loss'])))
epochs_small = list(range(0, len(xLSTM_small['test_loss'])))

plt.figure(figsize=(10, 6))
plt.plot(epochs_big, xLSTM_big["train_loss"], label=f'xLSTM - train_loss', color="g")
plt.plot(epochs_big, xLSTM_big["test_loss"], label=f'xLSTM - test_loss', linestyle='--', color="g")
plt.plot(epochs_small, LSTM_big["train_loss"], label=f'LSTM - train_loss', color = "r")
plt.plot(epochs_small, LSTM_big["test_loss"], label=f'LSTM - test_loss', linestyle='--', color = "r")


plt.title(f'Loss Comparison')
plt.xlabel('Epoch')
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()
```
